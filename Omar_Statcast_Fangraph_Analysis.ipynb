```python
#pull our analysis data for 2018
analysis_data_18 = batting_stats_2018[["Avg Launch Angle", "Avg EV(mph)", "Barrels%","wRC+"]]
```

```python
analysis_data_18.head()
```

```python
#define our independent and dependent variables. wRC+ will be our y
X = analysis_data_18.iloc[:,:-1].values
y = analysis_data_18.iloc[:,3].values
```

```python
#Next step is that we need to split our dataset into a training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
X.shape
```

```python
#fitting multiple lineaer regression to the training set
from sklearn.linear_model import LinearRegression
```

```python
#define regressor and then fit it to  our training set
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#predicting the Test set results
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
#building the optimal model using backward elimination
#this library doesn't include the X0 by default
#we'll need to manually add a column of 1 to our independent values
# import statsmodels.formula.api as sm
import statsmodels.regression.linear_model as sm
```

```python
X = np.append(arr = np.ones((448,1),dtype=int), values = X ,axis=1)
```

```python
X
```

```python
#optimal matrix of features, that are statistically significant for dependent variable
#we need to select our significance level to stay in the model (SL = 0.05)
#step 2: Fit the full model with all possible predictors
#need to create a new regressor object for the statsmodel class
X_opt = X[:,[0,1,2,3]]
regressor_OLS = sm.OLS(endog = y, exog= X_opt).fit()
```

```python
#step 3: Consider the predictor with the highest P-Vlaue. If P>SL, go to step 4, otherwise go to FIN
regressor_OLS.summary()
```

```python
#Step 4 we now need to remove highest p-value, X1, index1, and fit model again
X_opt = X[:,[0,2,3]]
regressor_OLS = sm.OLS(endog = y, exog= X_opt).fit()
```

```python
regressor_OLS.summary()
```
